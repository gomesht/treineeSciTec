```python
# Converter colunas de texto para o tipo 'category'
for col in ['PAIS', 'MORADOR', 'BLOCO','PAIS', 'ESTADO', 'CIDADE', 'OUTRA_CIDADE', 'BAIRRO_BH', 'REGIAO_BH', 'REGIONAL_ORIGEM','MOTIVO', 'GENERO', 'ESCOLARIDADE', 'ESTADO_CIVIL','PRETENDERIA_VIAJAR', 'PERNOITE_BH','HOSPEDAGEM', 'TRANSPORTE','MOTIVO_CARNAVAL','MOTIVO_OUTRO','PESSOAS_GASTOS','FONTE_CARNAVAL','FONTE_CARNAVAL', 'PARTICIPOU_ANOSANTERIORES','EXPECTATIVAS', 'AVAL_GERAL', 'AVAL_BANHEIROS', 'AVAL_SEGURANCA', 'AVAL_PRECOS', 'PROXIMA_EDICAO', 'MOTIVO_NAO','BLOCOS_DE_RUA', 'EVENTOS_PRIVADOS', 'PALCOS', 'CIRCUITO_GASTRONOMICO', 'CITY_TOUR', 'VISITA_CULTURAL', 'NAO_SABE', 'PRINCIPAL_PATROCINADOR','FAIXA_DE_RENDA', 'TRANSP_A_PE', 'TRANSP_APP', 'TRANSP_BICICLETA', 'TRANSP_CAR_AMIGOS_PARENTES', 'TRANSP_TAXI', 'TRANSP_OUTROS', 'TRANSP_VEICULO_PROPRIO', 'TRANSP_VEICULO_ALUGADO', 'TRANSP_ONIBUS_COLETIVO', 'TRANSP_NS', 'TRANSP_METRO' ]:
    df[col] = df[col].astype('category')
#Converter coluna numérica para tipo inteiro
for col in ['T_PERNOITE', 'DIAS_CARNAVAL']:
    df[col] = df[col].replace('nr_nao_sabe', 0).fillna(0).astype(int)
#Converter colunas de gastos para tipo float (AQUI ACHO QUE TEVE PROBLEMA APARECERAM VALORES MUITO ALTOS)
for col in ['GASTOS_PRETENDIDOS','ALIMENTACAO_GASTOS', 'TRANSPORTE_GASTOS', 'FANTASIAS_GASTOS', 'COMPRAS_GASTOS', 'INGRESSOS_GASTOS', 'OUTROS_GASTOS', 'HOSPEDAGEM_GASTOS', 'PASSEIOS_GASTOS', 'TOTAL_GASTOS', 'INDIVIDUAL_GASTOS', 'DIARIO_GASTOS']:
    df[col] = df[col].replace('nr_nao_sabe', 0)
    # Remove todos os pontos e espaços, depois troca vírgula por ponto
    df[col] = df[col].astype(str).str.replace('.', '', regex=False).str.replace(',', '.', regex=False).str.replace(' ', '')
    # Se ainda houver valores não numéricos, substitui por 0
    df[col] = pd.to_numeric(df[col], errors='coerce').fillna(0).astype(float)
#Converter coluna de data para tipo datetime
df['DT_VISITA'] = pd.to_datetime(df['DT_VISITA'], errors='coerce')
#Converter coluna de duração para tipo timedelta
df['TEMPO_TOTAL_ENTREV'] = pd.to_timedelta(df['TEMPO_TOTAL_ENTREV'], errors='coerce')
print(f"Uso de memória DEPOIS  (Real): {df.memory_usage(deep=True).sum() / 1024:.2f} KB")
print(f"Uso de memória DEPOIS  (Padrão): {df.memory_usage(deep=False).sum() / 1024:.2f} KB")
```

```python
fig, axes = plt.subplots(2, 2, figsize=(14, 10))
#histograma da idade
sns.histplot(df['IDADE'], bins=30, kde=True, ax=axes[0, 0])
axes[0, 0].set_title('Distribuição de Idade')
axes[0, 0].set_xlabel('Idade')
axes[0, 0].set_ylabel('Frequência')
#histograma da dias de carnaval
sns.histplot(df['DIAS_CARNAVAL'], bins=10, kde=True, ax=axes[0, 1])
axes[0, 1].set_title('Distribuição de Dias de Carnaval')  
axes[0, 1].set_xlabel('Dias de Carnaval')
axes[0, 1].set_ylabel('Frequência') 
#filtrar gastos 
df_gastos_filtrados = df[(df['GASTOS_PRETENDIDOS'] > 0) & (df['TOTAL_GASTOS'] > 0)]
#histograma dos gastos pretendidos
sns.histplot(df_gastos_filtrados['GASTOS_PRETENDIDOS'], bins=10, kde=True, ax=axes[1, 0])
axes[1, 0].set_title('Distribuição de Gastos Pretendidos')
axes[1, 0].set_xlabel('Gastos Pretendidos (R$)')
axes[1, 0].set_ylabel('Frequência') 

#histograma dos gastos totais removendo 0
sns.histplot(df_gastos_filtrados['TOTAL_GASTOS'], bins=10, kde=True, ax=axes[1, 1])
axes[1, 1].set_title('Distribuição de Gastos Totais')    
axes[1, 1].set_xlabel('Gastos Totais (R$)')
axes[1, 1].set_ylabel('Frequência')
plt.tight_layout()
plt.show()
```

```python
#boxplot da idade
plt.figure(figsize=(8, 6))  
sns.boxplot(y=df['IDADE'])
plt.title('Boxplot da Idade') 
plt.xlabel('Idade')
plt.show()

#boxplot das idades por gênero
plt.figure(figsize=(10, 6)) 
sns.boxplot(x=df['GENERO'], y=df['IDADE'])
plt.title('Boxplot da Idade por Gênero')
plt.xlabel('Gênero')
plt.ylabel('Idade') 
plt.show()
```

```python
#grafico de dispersão entre idade e dias de carnaval
plt.figure(figsize=(10, 6))
sns.scatterplot(x=df['DIAS_CARNAVAL'], y=df['IDADE'])
plt.title('Dispersão entre Idade e dias de Carnaval')
plt.ylabel('Idade')
plt.xlabel('dias de Carnaval')
plt.show()
#grafico de dispersão entre bloco e idade
plt.figure(figsize=(14, 6))
sns.scatterplot(x=df['BLOCO'], y=df['IDADE'])
plt.title('Dispersão entre Idade e Bloco')
plt.ylabel('Idade')
plt.xlabel('Bloco')
plt.xticks(rotation=60) 
plt.tight_layout()     
plt.show()
```

```python
#grafico dispersãpo idade dias de carnaval e bloco
sns.scatterplot(y='IDADE', x='DIAS_CARNAVAL', hue='GENERO', data=df, alpha=0.7, palette='Set2')
plt.title('Relação entre Idade e dias de carnaval por Gênero')
plt.ylabel('Idade')
plt.xlabel('Dias de Carnaval')
plt.legend(title='Gênero')
plt.show()
```

```python
#grafico de pizza dos blocos mais populares
block_counts = df['BLOCO'].value_counts()
plt.figure(figsize=(10, 10))
plt.pie(block_counts, labels=block_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribuição dos Blocos Mais Populares') 
plt.show()
```

```python
#grafico de media de idade por bloco
block_age_means = df.groupby('BLOCO')['IDADE'].mean().sort_values(ascending=False)
plt.figure(figsize=(14, 6))    
sns.barplot(x=block_age_means.index, y=block_age_means.values)
plt.title('Média de Idade por Bloco')
plt.ylabel('Média de Idade')
plt.xlabel('Bloco')
plt.xticks(rotation=60) 
plt.tight_layout() 
plt.show()
```
